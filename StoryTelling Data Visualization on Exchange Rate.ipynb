```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

## When we change two or more columns name change

```python
conv_rate.rename(columns = {'[US dollar ]' : 'US_Dollar' , 'Period\\Unit:' : 'Time'} , inplace = True)
```

```python
conv_rate['Time'] = pd.to_datetime(conv_rate['Time'])
```

```python
conv_rate.info()
```

```python
conv_rate["US_Dollar"]
```

```python
#Date comes in ascending order
conv_rate.sort_values('Time' , inplace = True) 
```

```python
conv_rate['Time'].head(10)
```

```python
#Need to take only 2 columns in dataset
euro_to_dollar = conv_rate[['Time' , 'US_Dollar']].copy()
euro_to_dollar
```

```python
euro_to_dollar['US_Dollar'].value_counts()
```

```python
euro_to_dollar['US_Dollar'].value_counts().head(50)
```

```python
#'- comes in US_Dollar data, maybe Holiday, so remove these rows'
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_Dollar'] != '-']
euro_to_dollar.info()
```

```python
#convert to float
euro_to_dollar['US_Dollar'] = euro_to_dollar['US_Dollar'].astype(float)
euro_to_dollar.info()
```

```python
plt.figure(figsize = (9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'] , euro_to_dollar['US_Dollar'])
plt.title('Original Values' , weight = 'bold')

for i, rolling_mean in zip ([2,3,4,5,6] , [7,30,50,100,365]): #first bracket graphs , #2nd bracket for days
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'] , euro_to_dollar['US_Dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean) , weight = 'bold')

plt.tight_layout() #adjust b/w subplots
plt.show() 
```

```python
import matplotlib.style as style
import matplotlib.pyplot as plt

style.use('fivethirtyeight')  # Apply 538 style

# Adding plot
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(fin_crisis_2006_2009['Time'], fin_crisis_2006_2009['rolling_mean_30Days'], linewidth=1, color='#A6D785')  
# Green color line plot

# Highlight the crisis year 2008
ax.plot(fin_crisis_2008['Time'], fin_crisis_2008['rolling_mean_30Days'], linewidth=3, color='#e23d28')

ax.set_xticklabels([])  # remove x-axis text

x = 0.02
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform=ax.transAxes)
    x += 0.23

print(plt.yticks())

ax.set_yticklabels([]) #remove text from y-axis

y= 0.07

for rate in ['1.2' , '1.3' , '1.4' , '1.5' , '1.6']:
    ax.text(-0.04 ,y , rate, alpha = 0.5 , fontsize = 11 , transform =plt.gca().transAxes)
    y += 0.23

#Adding title and sub-title

ax.text(-0.05 , 1.2 ,'Euro-USD rate peaked at "1.59 USD equals to 1 Euro" in 2008 financial year crisis.' ,\
        weight = 'bold', transform = plt.gca().transAxes)

ax.text(-0.05, 1.1 , 'Euro-USD exchange rate b/w 2006 to 2010.' , size =13 , transform = plt.gca().transAxes)

plt.show()
```

```python
import matplotlib.style as style
import matplotlib.pyplot as plt

style.use('fivethirtyeight')  # Apply 538 style

# Adding plot
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(fin_crisis_2006_2009['Time'], fin_crisis_2006_2009['rolling_mean_30Days'], linewidth=1, color='#A6D785')  
# Green color line plot

# Highlight the crisis year 2008
ax.plot(fin_crisis_2008['Time'], fin_crisis_2008['rolling_mean_30Days'], linewidth=3, color='#e23d28')

ax.set_xticklabels([])  # remove x-axis text

x = 0.02
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform=ax.transAxes)
    x += 0.23

ax.set_yticklabels([]) #remove text from y-axis

y= 0.07

for rate in ['1.2' , '1.3' , '1.4' , '1.5' , '1.6']:
    ax.text(-0.04 ,y , rate, alpha = 0.5 , fontsize = 11 , transform =plt.gca().transAxes)
    y += 0.23

#Adding title and sub-title

ax.text(-0.05 , 1.2 ,'Euro-USD rate peaked at "1.59 USD equals to 1 Euro" in 2008 financial year crisis.' ,\
        weight = 'bold', transform = plt.gca().transAxes)

ax.text(-0.05, 1.1 , 'Euro-USD exchange rate b/w 2006 to 2010.' , size =13 , transform = plt.gca().transAxes)

#Adding Signature 

ax.text(-0.05 , -0.25, '@TECHMA ZONE' + " "*80 + "Source:European Central Bank" , color ="#f0f0f0" \
        , bbox=dict(facecolor='#4d4d4d', alpha=0.5) ,transform = plt.gca().transAxes )
        
#Highlight the crisis start and end

ax.axvspan(xmin = pd.to_datetime('2008-04-1') , xmax = pd.to_datetime('2008-09-1') ,ymin = 0.09 , alpha = 0.3, \
           color = 'grey')

plt.show()
```

```python
import matplotlib.style as style
import matplotlib.pyplot as plt

style.use('fivethirtyeight')  # Apply 538 style

# Adding plot
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(corona_crisis_16_to_19['Time'], corona_crisis_16_to_19['rolling_mean_30Days'], linewidth=1, color='#A6D785')  
# Green color line plot

# Highlight the crisis year 2020
ax.plot(corona_crisis_20['Time'], corona_crisis_20['rolling_mean_30Days'], linewidth=3, color='#e23d28')

ax.set_xticklabels([])  # remove x-axis text

x = 0.02
for year in ['2016', '2017', '2018', '2019', '2020' , '2021']:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform=ax.transAxes)
    x += 0.18

ax.set_yticklabels([]) #remove text from y-axis

y= 0.07

for rate in ['1.05' , '1.10' , '1.15' , '1.20' , '1.25']:
    ax.text(-0.04 ,y , rate, alpha = 0.5 , fontsize = 11 , transform =plt.gca().transAxes)
    y += 0.23

#Adding title and sub-title

ax.text(-0.05 , 1.2 ,"Euro-USD rate peaked at 1.22 during 2022s Covid 19 Crisis" ,\
        weight = 'bold', transform = plt.gca().transAxes)

ax.text(-0.05, 1.1 , 'Euro-USD exchange rate b/w 2006 to 2010.' , size =13 , transform = plt.gca().transAxes)

#Adding Signature 

ax.text(-0.05 , -0.25, '@TECHMA ZONE' + " "*80 + "Source:European Central Bank" , color ="#f0f0f0" \
        , bbox=dict(facecolor='#4d4d4d', alpha=0.5) ,transform = plt.gca().transAxes )
        
#Highlight the crisis start and end



plt.show()


plt.show()
```

## Comparison three USA President, USD exchange rate compare to Euro

```python
bush_obama_trump = euro_to_dollar.copy(
    )[(euro_to_dollar['Time'].dt.year >= 2001) &
     (euro_to_dollar['Time'].dt.year < 2021)]
bush_obama_trump
```

```python
###all president years 2001 to 2021
bush_obama_trump = euro_to_dollar.copy(
    )[(euro_to_dollar['Time'].dt.year >= 2001) &
     (euro_to_dollar['Time'].dt.year <= 2021)]

#bush year 2001 to 2009
bush = bush_obama_trump.copy(
    )[bush_obama_trump['Time'].dt.year < 2009]

#obama period 2009 to 2017
obama = bush_obama_trump.copy(
    )[(bush_obama_trump['Time'].dt.year >= 2009) &
     (bush_obama_trump['Time'].dt.year < 2017)]

#trump period 2017 to 2021
trump = bush_obama_trump.copy(
    )[(bush_obama_trump['Time'].dt.year >= 2017) &
     (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
## Making Grids
## 538 style use
style.use('fivethirtyeight')

#add subplots
plt.figure(figsize = (12,6))

#Pattern 1
ax1 = plt.subplot(3,3,1) #row, col , index number
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)

#Pattern 2
ax4 = plt.subplot(3,1,2)
```

```python
##Compare 3 President performance in term of USD exchange rate againt EURO by graph
#add subplot
plt.figure(figsize = (14,8))

#Pattern 1
ax1 = plt.subplot(3,3,1) #row, col , index number
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)

#Pattern 2
ax4 = plt.subplot(3,1,2)

axes = [ax1 , ax2 , ax3 , ax4]

#changes all sub_plots
for ax in axes:
    ax.set_ylim(0.8 , 1.7)
    ax.set_yticks([1.0 , 1.2 , 1.4 , 1.6])
    ax.set_yticklabels(['1.0' , '1.2' , '1.4' , '1.6 $'] , alpha = 0.4)
    
#AX1: BUSH
ax1.plot(bush['Time'] , bush['rolling_mean_30Days'] , color = 'blue')
ax1.set_xticklabels(['', '2001', '' , '2003','','2005','', '2007','' , '2009'] , alpha= 0.3, size = 12)
ax1.text(0.09, 2.6 ,'G. BUSH' , fontsize = 15 , color= 'blue' , transform = plt.gca().transAxes)
ax1.text(0.09 , 2.45 , '(2001-2009)'  , alpha = 0.3 , transform = plt.gca().transAxes)

#AX2: OBAMA
ax2.plot(obama['Time'] , obama['rolling_mean_30Days'] , color = 'orange')
ax2.set_xticklabels(['', '2001', '' , '2003','','2005','', '2007','' , '2009'] , alpha= 0.3, size = 12)
ax2.text(0.44, 2.65 ,'B. OBAMA' , fontsize = 15 , color= 'orange' , transform = plt.gca().transAxes)
ax2.text(0.43, 2.5 , '(2009-2017)'  , alpha = 0.3 , transform = plt.gca().transAxes)

#AX3: Trump
ax3.plot(trump['Time'] , trump['rolling_mean_30Days'] , color = 'yellow' , alpha = 0.7)
ax3.set_xticklabels(['', '2001', '' , '2003','','2005','', '2007','' , '2009'] , alpha= 0.3, size = 12)
ax3.text(0.8, 2.65 ,'D. Trump' , fontsize = 15 , color= 'yellow' , alpha= 0.7 , transform = plt.gca().transAxes)
ax3.text(0.79, 2.5 , '(2017-2021)'  , alpha = 0.3 , transform = plt.gca().transAxes)

#AX4: All President performnce in one graph
ax4.plot(bush['Time'] , bush['rolling_mean_30Days'] , color = 'blue')
ax4.plot(obama['Time'] , obama['rolling_mean_30Days'] , color = 'orange')
ax4.plot(trump['Time'] , trump['rolling_mean_30Days'] , color = 'yellow')

#Add a Signature
ax4.text(-0.05 , -0.25, '@TECHMA ZONE' + " "*140 + "Source:European Central Bank" , color ="#f0f0f0" \
        , bbox=dict(facecolor='#4d4d4d', alpha=0.5) ,transform = plt.gca().transAxes )

plt.tight_layout()
plt.show()
```
